Now, to use visualization, import required libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import bokeh.plotting as bkp
from bokeh.io import show
```

***Line Plots:*** Line plots are useful for visualizing the relationship between two continuous variable. We can create line plots using Matplotlib or Seaborn.

```python
x=[1,2,3,4,5]
y=[10,12,15,18,20]
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('X-axis')
```

```python
plt.ylabel('Y-axis')
```

```python
plt.title('Line Plot')
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

### ***Bar Charts:*** Bar charts are ideal for comparing categorical data.We can create bar charts using Matplotlib or Seaborn.

```python
data = {'Category':['A','B','C','D'],
        'Value':[10,15,7,12]}
```

```python
sns.barplot(x='Category',y='Value',data=data)
```

```python
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart')
```

```python
sns.barplot(x='Category',y='Value',data=data)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart')
plt.show()
```

## ***Scatter Plots:*** Scatter plots are useful for visualizing the relationship between two continuous variables. We can create scatter plots using Matplotlib or Plotly.


```python
import pandas as pd
```

```python
data = pd.DataFrame({
    'X':[1,2,3,4,5],
    'Y':[10,12,15,18,20]
})
```

```python
fig = px.scatter(data)
fig.show()
```

## ***Histograms:*** Histograms are used to visualize the distribution of a single variable. We can create histograms using Matplotlib or Seaborn.

```python
data = [1,1,2,3,4,4,4,5,5,6,7,8,9]
```

```python
plt.hist(data, bins=5)
```

```python
plt.hist(data, bins=5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

# ***Interactive Visualizations:*** Plotly and Bokeh provide interactive visualization capabilities.

```python
x=[1,2,3,4,5]
y=[10,12,15,18,20]
bkp.output_notebook()
```

```python
p=bkp.figure(title='Interactive Line Plot',x_axis_label='X-axis',y_axis_label='Y-axis')
p.line(x,y)
show(p)
```
